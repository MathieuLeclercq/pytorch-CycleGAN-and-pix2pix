```python
import random 
import cv2 
from matplotlib import pyplot as plt 
import albumentations as A
```

```python
def visualize(image): 
    plt.figure(figsize=(10, 10)) 
    plt.axis('off') 
    plt.imshow(image)
```

```python
image = cv2.imread(r'C:\Users\M47h1\Documents\EiffelMedtech\pytorch_cyclegan\datasets\xray2drr\trainB\s0006_6_ap.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# visualize(image)
```

```python
transform = A.Compose([
    A.Rotate(limit=(-15,15)),
    A.HorizontalFlip(p=0.5),
    A.GridDistortion(p=0.9),
    A.Affine(scale=2),
    
    
])

```

```python
augmented_image = transform(image=image)['image']
visualize(augmented_image)
```
